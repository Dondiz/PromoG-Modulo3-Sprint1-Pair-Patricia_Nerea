```python
import pandas as pd
pd.set_option('display.max_columns', None)
```

```python
df_p1 = pd.read_csv("files/world-data-2023_part1.csv", index_col= 0)
df_p1.columns
```

```python
df_p2 = pd.read_csv("files/world-data-2023_part2.csv", index_col= 0)
df_p2.columns
```

```python
df_final = df_p1.merge(df_p2, how = "inner", left_on = "Country", right_on= "country")
df_final
```

Después de la unión de datos, tenemos dos columnas de "country". Elimina una de ellas.



```python
df_limpio = df_final.drop("country", axis = 1)
```

```python
df_limpio
```


Los nombres de las columnas no son homogeneos. Cambia los nombres de las columnas de tal forma que:

No tengan espacios.

Estén en minúscula.

No tengan paréntesis, es decir, quitar "(%)", "(Km2)".

Algunas columnas tiene "\n". Eliminalos de los nombres de las columnas.

Algunas columnas tienen ":". Eliminalos de los nombres de las columnas.



```python
df_espacios = df_limpio.columns.str.replace(" ", "")
df_limpio.columns = df_espacios
```

```python
df_limpio
```

```python
columnas_sin_espacios = [col.replace(" ", "") for col in df_limpio.columns]
df_limpio.columns = columnas_sin_espacios
```

```python
df_limpio["coordinates"]
```

```python
df_minusculas = df_espacios.str.lower()
df_limpio.columns = df_minusculas
```

```python
df_parentesis = df_minusculas.str.replace("\([^)]*\)", "", regex=True)
df_limpio.columns = df_parentesis
```

```python
df_salto = df_parentesis.str.replace("\n", "")
df_limpio.columns = df_salto
```

```python
df_corregido = df_salto.str.replace(":", "")
df_limpio.columns = df_corregido
```

```python
df_limpio
```

La columnas coordinates tiene la latitud y la longitud en una sola columna. Crea dos columnas nuevas, una con la longitud y otra con la latitud. Una vez hecho, elimina la columna de coordinates.



```python
df_limpio[['latitude', 'longitude']] = df_limpio['coordinates'].str.split(',', expand=True)
```

```python
df_limpio['coordinates'].str.split(',', expand=True)
```

```python
df_limpio.drop("coordinates", axis=1)
```

Las columnas unemployment_rate, total_tax_rate, tax_revenue, population_labor_force_participation, out_of_pocket_health_expenditure, gross_tertiary_education_enrollment, gross_primary_education_enrollment, forested_area, cpi_change, agricultural_land tienen "%". Elimina los "%" de los valores de las columnas.

```python
columnas= [
    "unemploymentrate",
    "totaltaxrate",
    "taxrevenue",
    "populationlaborforceparticipation",
    "outofpockethealthexpenditure",
    "grosstertiaryeducationenrollment",
    "grossprimaryeducationenrollment",
    "forestedarea",
    "cpichange",
    "agriculturalland",
]
```

```python
for columna in columnas:
    df_limpio[columna] = df_limpio[columna].str.replace("%", "")
```

```python
df_limpio
```

```python
df_limpio.columns
```

```python
columnas2 = ["gasolineprice", "gdp", "minimumwage"]
```

```python
for columna in columnas2:
    df_limpio[columna] = df_limpio[columna].str.replace("$", "", regex=True)
```

```python
df_limpio
```

```python
df_limpio.to_csv('files/df_limpio.csv', index=False)
```
